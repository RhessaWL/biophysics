## 2. Population Coding

Simulate a population of direction-selective neurons with Poisson spike statistics, Gaussian-like tuning of width σ = 60º and maximal response of 100 spikes/s, which have equidistant preferred orientations sampling the range from 0º to 360º with a sampling interval of 10º.

Calculate the responses to the presentation of a 204º stimulus as spike counts in 500 ms by drawing from a Poisson distribution. From the responses, estimate the stimulus orientation by 

1) the preferred orientation of the neuron with the largest firing rate (Winner-Take-All approach)
2) the Population Vector
3) the center of mass of the response distribution
4) fitting a Gaussian with width corresponding to the expected mean response distribution.

_Step 1:_
Start by defining an array of the preferred directions of the neurons. This is the only parameter that varies across the neurons, so this array represents the population.

_Step 2:_
Define the tuning function. It should take the maximal response, the tuning width, the preferred angle of the neuron, and the stimulus angle as arguments and return the mean response for that stimulus angle.

We have a periodic stimulus variable (because we want an example where we can use the Population Vector). To approximate a Gaussian on a periodic scale we use a von Mises function,

$$ f(s) = \frac{1}{2\pi i_0(\kappa)} e^{( \kappa \cos(s - s_0) )}$$
where $i_0$ is the modified Bessel function of order zero. The relation between the width $\sigma$ of the approximated Gaussian and the concentration parameter $\kappa$ is $\kappa \approx \frac{1}{\sigma^2}$.

Note that you can omit the normalizing factor - we don't want to represent probabilities but just use the shape of the functions as tuning curves.
Therefore, it is easier to just multiply the exponential with the maximal firing rate (or maximal spike count for that matter), so that the function directly yields the mean response.

Plot the tuning curve for the neuron with preferred angle $s_0 = 180º$.
